```python
import pandas as pd
import numpy as np
```

```python
journey9 = pd.read_hdf('journey9.hdf', 'journey9')
journey9.tail().T
```

```python
train = journey9[journey9["subscription_expiry_month"] < "2020-11-30"]
test = journey9[journey9["subscription_expiry_month"] >= "2020-11-30"]
```

```python
HEAD = train.head(1000)
```

```python
ID_COL = "expiry_principal_subscription_id"
TARGET_COL = "0_month_renewal"
REM_COLS = ["1_month_renewal"]

NUM_COLS = [
            NUM_COL 
            for NUM_COL in HEAD.select_dtypes('number').columns \
            if NUM_COL not in REM_COLS +  [TARGET_COL , ID_COL]
           ]
OBJ_COLS = list(HEAD.select_dtypes('object').columns)
```

```python
journey9[NUM_COLS] = journey9[NUM_COLS].fillna(0)
```

```python
df = journey9.groupby([
    "subscription_expiry_month", "subscription_expiry_transaction_source"
])[NUM_COLS + ["0_month_renewal"]].mean()

pd.pivot_table(df,
               values=NUM_COLS + ["0_month_renewal"],
               index=["subscription_expiry_month"],
               columns=["subscription_expiry_transaction_source"],
               aggfunc=np.sum)
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.pipeline import Pipeline, FeatureUnion
from sklearn.compose import ColumnTransformer
from sklearn.base import TransformerMixin, BaseEstimator
from sklearn.impute import SimpleImputer
from sklearn import set_config
set_config(display='diagram')

from utils import DFImputer, DFLabelEncoder
```

```python

num_processor = Pipeline([
    ('mean_imputer', DFImputer(strategy='constant', fill_value=0)),
])

obj_processor = Pipeline([('str_imputer', DFImputer(strategy='most_frequent')),
                          ('dummify', DFLabelEncoder())])

preprocess_pipeline = ColumnTransformer(
    [('num_proc', num_processor, NUM_COLS),
     ('obj_proc', obj_processor, OBJ_COLS)],
    remainder='passthrough'
)

```

```python
TRF_DF = pd.DataFrame(preprocess_pipeline.fit_transform(HEAD), columns=NUM_COLS +
             OBJ_COLS + [col for col in HEAD.columns if col not in NUM_COLS + OBJ_COLS])
```
